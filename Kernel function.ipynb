```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
epsilon=0.25
chi=2.0
x0=0
xf=5
dx=0.001
```

```python
x=np.arange(x0,xf,dx)
w=[]
for xi in x:
    wi=np.exp(-xi**2)-epsilon*np.exp(-(xi/chi)**2)
    w.append(wi)
plt.plot(x,w)
plt.axhline(y=0, color='k', ls='dashed', lw=1)
plt.show()
```
